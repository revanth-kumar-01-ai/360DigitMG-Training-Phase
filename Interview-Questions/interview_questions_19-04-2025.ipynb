**1. What is deep learning, and how does it differ from traditional machine learning?** 

## Deep learning 

Deep learning is a subset of machine learning

That uses neural network with many layers (like brain) to learn  for big data

And it solve complex problem like images voice etc 

### Difference 

**Traditional Machine learning**

Traditional Machine Learning needs manual feature selection and works well with small to medium-sized data. It uses simple models like SVM and Decision Trees. These models are fast to train, require less computation power, and usually run well on CPUs üíª. Traditional ML is best suited for structured, tabular data.

**Deep Learning**

Deep Learning does feature engineering automatically, so no need to hand-pick features. It can handle large amounts of data and uses complex neural networks. Training is slow but very powerful. It needs high computation power and works best with GPU (not just CPU üíª). Deep learning is great for images, speech, and video.



**2. How is deep learning related to artificial neural networks?**

Deep Learning is a subset of Artificial Intelligence (AI), and it's built on Artificial Neural Networks (ANNs).

An ANN is inspired by how the human brain works. It consists of layers of neurons (nodes) that process and learn from data. In Deep Learning, these ANNs are made much deeper (more layers) to solve complex tasks.

Deep Learning uses these deep neural networks to automatically find patterns in large datasets, such as images, speech, or text. The more layers in a network, the better it can learn complex features on its own, which is why it's so powerful for tasks like image recognition, speech recognition üéß, and language translation.

**5. Can you explain the differences between supervised, unsupervised, and reinforcement learning in the context of deep learning?**

##### Supervised Learning

- Model learns from labeled data (input + correct output).

##### Unsupervised Learning

-  Model learns from unlabeled data and tries to find patterns or structures.

##### Reinforcement Learning

- Model learns by interacting with an environment, receiving rewards or punishments based on its actions.

**6. What are the main components of a neural network?**

- 1. **Input Layer**

    * Takes the raw data (like image, text, numbers)
    * Each neuron = 1 feature 

- 2. **Hidden Layers**

    * Do the real processing with neurons
    * Each neuron = weight √ó input + bias 
    * Then passes through activation 

- 3. **Weights**

    * Controls the importance of inputs 
    * Gets updated during training

- 4. **Bias**

    * Extra value added to shift output
    * Helps model learn better 

- 5. **Activation Functions**

    * Adds non-linearity (like ReLU, Sigmoid)
    * Makes model powerful 

- 6. **Output Layer**

    * Final prediction (class or number)
    * Depends on task (e.g., softmax, linear) 


**7. What is backpropagation, and how does it help train neural networks?**

### Backpropagation = "Backward learning"

It helps the neural network learn from mistakes by updating the weights! 

### How it works:

1. **Forward Pass** ‚û°Ô∏è

   * Input goes through the network
   * Get output & calculate error (loss) 

2. **Backward Pass** ‚¨ÖÔ∏è

   * Use **backpropagation** to send error backward
   * Adjust weights using gradients (via **gradient descent**) 

3. **Repeat** for many epochs üîÅ

### üéØ Goal:

Reduce error by updating weights ‚Äî so the model becomes smarter! 




**8. What is gradient descent, and how does it work in deep learning?** 

##### What is Gradient Descent?

It‚Äôs a method to **update weights** in a neural network to **reduce the error (loss)** üìâ

### How it works:

1. Start with random weights 
2. Do a forward pass 
3. Calculate error (loss) 
4. Backpropagate the error ‚¨Ö
5. **Use gradients** to adjust weights in the direction that **reduces the loss** 
6. Repeat this for many epochs 

### üìå Simple Formula:

```
new_weight = old_weight - learning_rate √ó gradient
```

This helps the model learn better step by step 

Easy and powerful bro! 


**9. What is the difference between stochastic gradient descent (SGD) and batch gradient descent?**

### **Batch Gradient Descent**

* Uses **all data** to compute gradient 
* More accurate update 
* Slower ‚è≥

### **Stochastic Gradient Descent (SGD)**

* Uses **1 random sample** at a time 
* Faster üí®
* More noise/shaky update ‚ö°

### In Short:

| Type      | Data Used | Speed üöÄ | Accuracy üéØ |
| --------- | --------- | -------- | ----------- |
| **Batch** | All data  | Slow     | Stable      |
| **SGD**   | 1 sample  | Fast     | Noisy       |

Both are used in deep learning training üí™üî•


**10. What is the role of the learning rate in training deep learning models?**

### üéØ **Learning Rate** = Step size for learning

It controls **how much to change weights** during training after each update.

### üß† Role:

* **Small LR** üê¢: Slow training, but more accurate
* **Big LR** üöÄ: Fast training, but can overshoot or miss the best point
* **Just right LR** ‚úÖ: Fast + stable learning

### üìå Tip:

Pick a good learning rate to balance **speed vs accuracy** ‚öñÔ∏è
